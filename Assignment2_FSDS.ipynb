ANSWER- The 'Break' and 'Continue' both statement are used to skip the iteration of a loop.
       The 'Break' statement terminates the whole iteration of a loop and while 'Continue' skips the current iteration.
       Let's understand with an example:
              "Break statement"-> for i in range(4):
                                       for j in range(4):
